```python
df=df.drop(['S.No.'],axis=1) # axis=1 means the column
df.info()
```

### Feature Engineering

```python
### Creating a new variable called car_age
```

```python
from datetime import date
date.today().year
df['Car_Age']=date.today().year-df['Year']
df.head(5)
```

```python
# Split function-splits the word
```

```python
df['Brand']=df.Name.str.split().str.get(0)
```

```python
df['Model']=df.Name.str.split().str.get(1)+df.Name.str.split().str.get(2)
```

```python
df[['Name','Model','Brand']]
```

### Data Cleaning/Data Wrangling

```python
print(df.Brand.unique())
print(df.Brand.nunique())
```

```python
# ISUZU ,mini look wrong in this particular data
```

```python
searchfor=['ISUZU','Isuzu','Mini','Land']
df[df.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
df['Brand'].replace({'ISUZU':'Isuzu','Mini':'Mini Cooper','Land':'Land Rover'},inplace=True)
```

```python
df['Mileage']=df.Mileage.str.split().str.get(0)   # Gets the particular value from the data eg: 26.6 is one and kmpl is another
df['Mileage']
```

```python
df['Power']=df.Power.str.split().str.get(0)
df['Power']
```

```python
df['Mileage']=df['Mileage'].fillna(0).astype('float')
df['Mileage']
```

```python
df['Power']=df['Power'].fillna(0).astype('float')    # ERROR-----Bcz it has 46 null values
df['Power']
```

```python
df['Power'].replace("null",'0',inplace=True)
df['Power']
```

### Observation

```python
### Seperating Numerical and Categorical variables
```

```python
cat_cols=df.select_dtypes(include='object').columns
num_cols=df.select_dtypes(include=np.number).columns.tolist()
print("Categorical column list: ",cat_cols)
print("Numerical column list: ",num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew: ',round(df[col].skew(),2))            # skew- Distribution of the data (right skewness or left skewness)   
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)          #  1 row, 2 columns 1st position  (LHS- Histogram)
    df[col].hist(grid=False)
    plt.ylabel('Count')
    plt.subplot(1,2,2)          # 1 row 2 columns 2nd position    ()
    sns.boxplot(x=df[col]) 
    plt.show
```

```python
fig,axes=plt.subplots(3,2,figsize=(18,18))
fig.suptitle('Bar plot for all catrgorical variables in the datasets')
sns.countplot(ax=axes[0,0],x='Fuel_Type',data=df,color='cyan',
             order=df['Fuel_Type'].value_counts().index);
sns.countplot(ax=axes[0,1],x='Transmission',data=df,color='red',
             order=df['Transmission'].value_counts().index);
sns.countplot(ax=axes[1,1],x='Location',data=df,color='blue',
             order=df['Location'].value_counts().index);
sns.countplot(ax=axes[1,0],x='Owner_Type',data=df,color='black',
             order=df['Owner_Type'].value_counts().index);
sns.countplot(ax=axes[2,1],x='Model',data=df,color='green',
             order=df['Model'].head(20).value_counts().index);
sns.countplot(ax=axes[2,0],x='Brand',data=df,color='purple',
             order=df['Brand'].head(20).value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```
